```python
kdown = pd.DataFrame(df_output.BEERS.GlobalRad.loc[1])

def melt_df(df, var, kdown):
   df = df[var]
   # Add kdown and day/night columns to df
   df['kdown'] = kdown['GlobalRad']
   df['day_night'] = df['kdown'].apply(lambda x: 'day' if x >= 0.5 else 'night')
   
   # Select columns 1-15 and melt the DataFrame
   value_vars = df.columns[:15].tolist()
   melted_df = df.reset_index().melt(id_vars=['datetime', 'day_night'], value_vars=value_vars, var_name='grid', value_name=var)
   
   return melted_df
```

```python
df_s = df_output.SUEWS

df_s.eval('AE=QN+QF-QS', inplace=True)
df_s.eval('Br=QE/QH', inplace=True)
df_s.eval('EF=QE/(QH+QE)', inplace=True)
```

```python
grid = 11

ss = '2018-06-04'

# Plot QE and QH on the left y-axis
fig, ax1 = plt.subplots()

var = 'Br'

ax1.plot(df_s.loc[grid].loc[ss].index, df_s.loc[grid].loc[ss, 'QE'], label='QE', color='b')
ax1.plot(df_s.loc[grid].loc[ss].index, df_s.loc[grid].loc[ss, 'QH'], label='QH', color='g')
ax1.set_xlabel('Datetime')
ax1.set_ylabel('QE and QH')
ax1.legend(loc='upper left')

# Create a second y-axis to plot Br
ax2 = ax1.twinx()
ax2.plot(df_s.loc[grid].loc[ss].index, df_s.loc[grid].loc[ss, var], label=var, color='r')
ax2.set_ylabel(var)
ax2.legend(loc='upper right')

ax1.set_ylabel('W/m')
ax2.set_ylabel('W/m')
```

```python
fig, ax = plt.subplots(1, 3, figsize=(20, 5))
import seaborn as sns
# Ensure the grids are sorted

var = 'QE'
sns.boxplot(data=melt_df(df_s, var, kdown), x='grid', y=var, hue ='day_night' ,ax=ax[0],showfliers=False)
ax[0].set_title(var)
ax[0].set_ylabel('W/m')

var = 'QH'
sns.boxplot(data=melt_df(df_s, var, kdown), x='grid', y=var, hue ='day_night' ,ax=ax[1],showfliers=False)
ax[1].set_title(var)
ax[1].set_ylabel('W/m')

var = 'Br'
sns.boxplot(data=melt_df(df_s, var, kdown), x='grid', y=var ,ax=ax[2],showfliers=False)
ax[2].set_title(var)
ax[2].set_ylabel('W/m')


for axs in [0,1,2]:
    ax[axs].axhline(0, color = 'grey', alpha = 0.5, linestyle= '--')
```
