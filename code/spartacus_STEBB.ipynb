| Storageheatmethod | (1) "ΔQS modelled using the objective hysteresis model (OHM) [Grimmond et al., 1991] using parameters specified for each surface type." | (5) "EHC Module" |

Zenith angle not accounted for in albedo calculation." | (1003) "Q* modelled with SPARTACUS-Surface (SS) but with L↓ modelled as in 3. Experimental in this version." |
