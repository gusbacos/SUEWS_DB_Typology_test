```python


def melt_df(df, var, kdown):
   df = df[var]
   # Add kdown and day/night columns to df
   df['kdown'] = kdown['GlobalRad']
   df['day_night'] = df['kdown'].apply(lambda x: 'day' if x >= 0.5 else 'night')
   
   # Select columns 1-15 and melt the DataFrame
   value_vars = df.columns[:15].tolist()
   melted_df = df.reset_index().melt(id_vars=['datetime', 'day_night'], value_vars=value_vars, var_name='grid', value_name=var)
   
   return melted_df

```
